## Падение тела

```python
var("y1 y2 y3 y4 t")
alpha = pi/4
v0 = 10.0
k = 0.01
g = 9.81
tmax = 1.41


equation = [y2, -k * y2 * sqrt(y2 ^ 2 + y4 ^ 2), y4, -k * y4 * sqrt(y2 ^ 2 + y4 ^ 2) - g]
solution = desolve_system_rk4(equation, [y1, y2, y3, y4], ics = [0, 0, v0 * cos(alpha), 0, v0 * sin(alpha)], ivar = t, step = 0.01, end_points = tmax)

lines = plot([])

for i in range(len(solution) - 1):
    lines += line([[solution[i][1], solution[i][3]], [solution[i + 1][1], solution[i + 1][3]]])

show(lines)
```
